```python
# Why we use numpy if we have list
# Because List is very slow in comparison to Numpy
# why Numpy is faster ?
# Because numpy is uses Fixed type.
# for example we have a array in which a element is 5.
# as we know that 5 is stored in binary format : -00000101
# Numpy gives us flexibilty to store this value in: - int 8 = 0000101 or 
# int 16 or int 32 (00000000-00000000-00000000-00000101) 
# according to requirement.
# But in case of list it is built in binary so:-
# 1) size 
# 2) reference count
# 3) object type
# 4) object value
# so in lits these four parameter have there value so it will require more
# space in comparison to numpy.
# Conclusion
# Numpy is faster to read as less bytes of memory
# Numpy : -No type checking when iterating through object.
# list : -Type checking for int,float in every iteration.
# another reason is : - Contiguous memory.
# list :- list is kind of scatter around.It means memory block are scatter(spreaded around).
# Basically all memory are not in particular position and to catch this computer have to 
# bounce arround every memory. Therfore it is slow
# numpy :- but numpy uses the contiguous memory it means memory are not spreaded out
# they are in particular position(like we have 8 memeory block and they are staggerd in
# fix memory )

```

```python
print(a[0,]) # all the element of 1 rows
print()
print(a[0,1:5]) # 1 row and from 2(1) column to 4(5) column as 5 is not included
print()
print(a[0,1:5:2]) #a[all element of 0(1) row,column1:(to)column5(4):(stepsize)of 2]
print()
print(a[0,5:1:-1]) #negative index starting from 5(4) to 1
```

# Carefull while you doing copying


```python
a = np.array([1,2,3])
b = a
print(b) #check output

b[0] = 100
print(b)
print(a) #Now a is also changed as b is changed.
# this happened because as we know that the concept of list it points teh data same here 
# it is also pointing the data 
# nOw to prevent this we can use .copy 
c = np.array([1,2,3])
print(c)
d = c.copy()
d[0] = 99
print(d)
print(c) # value of c is not changed as here we used copy.
```

# Recoganizing Arrays


```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
print(before.shape)
print()
# Now we want to change the shape
after = before.reshape((8,1))
print(after)
print(after.shape)
# as 2 * 4 = 8 reshape should be in that order only otherwise it create an error
# print(before.reshape(3,3))
# ValueError: cannot reshape array of size 8 into shape (3,3)
```

```python
# Vertically stacking..of vectors..
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

print(np.stack([v1,v2]))
# we can stack multiple time also
print()
print(np.vstack([v1,v1,v2,v2]))

#Horizontal stacking..
h1 = np.ones((2,4))
h2 = np.zeros((2,4))
print()
print(np.hstack([h1,h2]))
```
